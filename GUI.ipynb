```python
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np


def f(k):
    x = np.linspace(0, 20, num=1000)
    plt.xlim(0, 20)
    plt.ylim(-5, 5)
    y=0;
    for n in range(1, k+1, 1):
        y = y + (-1)**(n-1)*2/n*np.sin(n*x)

    plt.title(r'$y=\sum_{n=1}^k(-1)^{n-1}* \frac{2}{n}*\sin(nx)$', fontproperties=fontprop)
    plt.plot(x, y)
    plt.show()

interact(f, k=(1,20,1) )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy.fftpack as spfft

def f(k):
    plt.xlim(0, 5)
    plt.ylim(0, 1.2)
    size = 500
    x = np.linspace(0, 5, size)
    y = 0.5+0.5*signal.square(2*np.pi * x)
    y[x>k]=0
    plt.plot(x, y)
    plt.show()
    freq = np.linspace(0, len(y), len(y))
    Y = spfft.fft(y) 
    plt.plot(freq[:len(y)//2+1], np.abs(Y[:len(y)//2+1]))

interact(f, k=(1,5,1) )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy.fftpack as spfft

def f(k):
    plt.xlim(0, 5)
    plt.ylim(-1.2, 1.2)
    size = 100
    fs = 5
    x = np.linspace(0, 5, size)
    y = np.sin(2 * np.pi * k* x / fs)
    plt.plot(x, y)
    plt.show()
    freq = np.linspace(0, len(y), len(y))
    Y = spfft.fft(y) 
    plt.xlim(0, 10)
    #plt.ylim(-1.2, 1.2)
    plt.plot(freq[:len(y)//2+1], np.abs(Y[:len(y)//2+1]))

interact(f, k=(1,10, 1) )
```
